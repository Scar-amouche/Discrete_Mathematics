# 1. Proving Identities 

$\textbf{A.}$ A\B = A $\cap$ $B^{c}$ is $\textbf{TRUE}$<br>
 if A\B={x | x $\in$ A $\land$ x $\notin$ B}<br>
 if A$\cap$ $B^{c}$={x | x $\in$ A $\land$ x $\notin$ B}<br>
 $\textbf{Example:}$<br>
 A={1,2,3}<br>
 B={3,4,5}<br>
 A\B={1,2}<br>
 A$\cap$ $B^{c}$={1,2}<br>
 
$\textbf{B.}$ B = (B $\cap$ A) $\cup$ (B $\cap$ $A^{c}$) is $\textbf{TRUE}$ <br>
 if B $\cap$ A={x | x $\in$ B $\land$ x $\in$ A} <br>
 if B $\cap$ $A^{c}$={x | x $\in$ B $\land$ x $\notin$ A} and;<br>
 if (B $\cap$ A) $\cup$ (B $\cap$ $A^{c}$) = {x | x $\in$ B $\land$ x $\in$ A}$\land${x | x $\in$ B $\land$ x $\notin$ A}<br>
 $\textbf{Example:}$<br>
 A={1,2,3}<br>
 B={3,4,5}<br>
 B $\cap$ A={3}<br>
 B $\cap$ $A^{c}$={4,5}<br>
 (B $\cap$ A) $\cup$ (B $\cap$ $A^{c}$)={3,4,5}<br>
 
$\textbf{C.}$ B\A = B $\cap$ $A^{c}$is $\textbf{TRUE}$ <br>
 if B\A={x | x $\in$ B $\land$ x $\notin$ A}<br>
 if B $\cap$ $A^{c}$={x | x $\in$ B $\land$ x $\notin$ A}<br>
 $\textbf{Example:}$<br>
 A={1,2,3}<br>
 B={3,4,5}<br>
 B\A={4,5}<br>
 B $\cap$ $A^{c}$={4,5}<br>
 
$\textbf{D.}$ A $\cup$ B = A $\cup$ (B $\cap$ $A^{c}$)is $\textbf{TRUE}$ <br>
 if A $\cup$ B = {x | x $\in$ A $\land$ x $\in$ B}<br>
 if A $\cup$ (B $\cap$ $A^{c}$) = {x | x $\in$ B $\lor$ x $\in$ A}=$A^{c}$<br>
 $\textbf{Example:}$<br>
 A={1,2,3}<br>
 B={3,4,5}<br>
 A $\cup$ B = {1,2,3,4,5}<br>
 B $\cap$ $A^{c}$ = {4,5}<br>
 A $\cup$ (B $\cap$ $A^{c}$) = {1,2,3,4,5}<br>

A_{i}^{c})^{c}$ = $\cap_{i=1}^{\infty}A_{i}$<br>
$(\cup_{i=1}^{\infty} \textbf{A}_{i}^{c} = (\textbf{U})^{c} = \emptyset$<br>
$\cap_{i=1}^{\infty}\textbf{A}_{i} = \emptyset$ the intersection of sets can be nothing, but $\cap_{i=1}^{\infty}\textbf{A}_{i} = \emptyset$ 

$\textbf{F.}$ |A $\cup$ B| = |A| + |B| - |A $\cap$ B|is $\textbf{TRUE}$ <br>
 It's the result of inclusion-exclusion principle<br>
 $\textbf{Example:}$<br>
 A={1,2,3}<br>
 B={3,4,5}<br>
 |A|=3<br>
 |B|=3<br>
 if we simply add them then there would be 1 repeating element<br>
 |A $\cap$ B|=1<br>
 |A $\cup$ B|=5<br>
 |A $\cup$ B| = |A| + |B| - |A $\cap$ B|<br>
       5      =  3  +  3  -       1<br>
       5      =  5<br>
